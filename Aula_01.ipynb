A relação entre o preço de venda "x" e o lucro "L" de uma determinada mercadoria é L(x)=-48x² + 12000x.
Calcule o preço que máximiza o lucro e o respectivo lucro máximo? **negrito**

```python
from sympy import *
x, l= symbols('x l')
l= -48*x**2+12000*x
df= diff(l, x)
d2f=diff(l, x, 2)
p= solve(Eq(df, 0))
ds= d2f.subs(x, p[0])
LM= l.subs(x, p[0])
print(f'Preço ótimo: R${p[0]}')
print(f'Derivaa segunda: {ds}')
print(f'Lucro máximo: R${LM}')
```

DESRIÇÂO MUITO GRANDE.
Consedere a função F(x,y)= 2x+y²+x².  Obtenha e classifique os respectiivos pontos criticos.

```python
from sympy import *
x, y, f = symbols ('x y f')
f= 2*x + y**2 + x**2
fx= diff(f, x) #1º derivada em relação a x
fy= diff(f, y) #1º derivada em relação a y
fxx= diff(f, x, 2)  #2º derivada em relação x
fyy= diff(f, y, 2)  #2º derivada em relação y
fxy= diff(fx, y)   # derivada em relação a y da função fx
fyx= diff(fy, x)   #derivada em relação e x da função fy
px= solve(Eq(fx, 0))
py= solve(Eq(fy, 0))
fxxp= fxx.subs({x:px[0], y:py[0]})
fyyp= fyy.subs({x:px[0], y:py[0]})
fxyp= fxy.subs({x:px[0], y:py[0]})
fyxp= fyx.subs({x:px[0], y:py[0]})
D= fxxp*fyyp - fxyp*fyxp
print(f'Solução: P ({px[0]}, {py[0]})')
print(f'Determinante: {D}')
print(f'Derivada segunda em relação a x: {fxxp}')
```

Calcule a área abaixoda função f(x)= x3-x no intevalo [2, 4] e faça o respectivo gráfico.

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
x, f= symbols ('x f')
f= x**3-x
A= integrate(f, (x, 2, 4))
x= np.linspace(1.5, 4.5, 1000)
f= x**3-x
plt.plot(x, f, color='red')
plt.axhline (color='red')
plt.fill_between (x, f, where= [(x>2) and (x<4) for x in x], color= 'blue')
plt.show()
print(f'Àrea:{A:.2f}')
```

Calcule a área entre as funções f(x)= x+2 e G(x)= 3/x no intervalo [1, 5]. Faça o gráfico?

```python
import matplotlib.pyplot as plt
from sympy import *
import numpy as np
x, f, g= symbols('x f g')
f= x+2
g= 3/x
A= integrate((f-g), (x, 1, 5))
x= np.linspace(0.5, 5.5, 1000)
f= x+2
g= 3/x
plt.plot (x, f, color='blue')
plt.plot (x, g, color='red')
plt.axhline(color='black')
plt.fill_between (x, f, g, where=[(x>1) and (x<5) for x in x], color='cyan')
plt.show()
print(f'Àrea: {A:.2f}')
```
