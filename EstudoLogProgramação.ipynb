Esse exercício pede três passos:

criar uma string
pegar metade dela
mostrar os dois últimos caracteres dessa metade

Vou te mostrar um exemplo em Python (se for outra linguagem, me fala que adapto):

# 1. Criando a string
frase = "Aprendendo programação"

# 2. Pegando a metade da string
metade = frase[:len(frase)//2]

# 3. Pegando os dois últimos caracteres da metade
resultado = metade[-2:]

# Exibindo o resultado
print(resultado)
Explicação rápida:
len(frase)//2 → pega o meio da string
frase[:...] → pega da posição inicial até o meio
[-2:] → pega os dois últimos caracteres

Faça um algoritmo que receba três valores, representando os lados de um triângulosfornecido pelo usuário. Verifique se os valores formam um triângulo e classifiqueos.
a)Equilatero ( três lados iguais)
b)Isosceles (dois lados iguais)
c)Escaleno (três lados diferente)

Lembre-se que para formar um triângulo nenhum dos lados pode ser igual a zero, e um lado não pode ser moir que a soma dos dois lados.


```python
a = int( input ('Digiteo o valor do 1º triângulo'))
b = int( input ('Digiteo o valor do 2º triângulo'))
c = int( input ('Digiteo o valor do 3º triângulo'))
if ((a>0 and b>0 and c>0) and (a+b>c and a+c>b and b+c>a)):
  #Se chegou até aqui, é porque o triânglo é válido.
  if(a!=b and a !=c and b !=c):
    print ('triâgulo escaleno')
  else:
    if(a==b and b==c):
      print('Triângulo equilatero')
    else:
      print ('Triângulo isosceles')
else:
  print('Ao menos um dos valores indicado ão servem para formar um retângulo')
```

Escreva um algoritmo que leia dois valores numéricos e que pergunte ao usuário qual operação deseja realizar: Adição (+), subtração (-), Multiplicação (*) ou Divisão (/). Exiba na tela o resultado desejado.

```python
print('Calculadora')
print(' + Adição')
print ('- Subtração')
print('* Multiplicação')
print('/ Divisão')
print('Precione qualquer tecla para sair')
op=input('Qual operaçõ deseja realizar')
x=int(input('digite o 1º valor'))
y=int(input('Digiteo 2º valor'))
if(op=="+"):
  res= x+y
  print(f'res:{x} + {y} = {res}')
elif (op=='-'):
  res=x - y
  print(f'Resultado: {x} - {y} = {res}')
elif (op=='*'):
  res=x * y
  print(f'Resultado: {x} * {y} = {res}')
elif (op=='/'):
  res=x / y
  print(f'Resultado: {x} / {y} = {res}')
else:
  print('Encerando o programa')
```
