```python
def calcular_venda():
    print("--- Sistema de Vendas ---")
    try:
        # Entrada de dados
        produto = input("Nome do produto: ")
        preco = float(input("Preço unitário: R$ "))
        quantidade = int(input("Quantidade: "))

        total_sem_desconto = preco * quantidade

        # Aplicação de desconto condicional (Ex: 10% se total > 100)
        percentual = 0
        if total_sem_desconto > 100:
            percentual = 10

        valor_desconto = total_sem_desconto * (percentual / 100)
        total_com_desconto = total_sem_desconto - valor_desconto

        # Exibição
        print("-" * 30)
        print(f"Resumo: {quantidade}x {produto}")
        print(f"Total sem desconto: R$ {total_sem_desconto:.2f}")
        if percentual > 0:
            print(f"Desconto aplicado ({percentual}%): R$ {valor_desconto:.2f}")
        print(f"Valor Final: R$ {total_com_desconto:.2f}")
        print("-" * 30)

    except ValueError:
        print("Erro: Entrada inválida. Use números para preço e quantidade.")

# Executar o programa
calcular_venda()
```

```python
def solicitar_valor_maximo(valor_limite):
    while True:
        entrada = input(f"Digite um valor (máximo {valor_limite}): ")

        try:
            # Tenta converter a entrada para float (aceita decimais)
            valor = float(entrada)

            # Verifica se o valor é maior que o limite
            if valor > valor_limite:
                print(f"Erro: O valor {valor} é maior que o limite permitido de {valor_limite}. Tente novamente.")
            else:
                print(f"Valor {valor} aceito!")
                return valor # Sai do loop e retorna o valor válido

        except ValueError:
            # Captura erro se o usuário digitar letras ou símbolos inválidos
            print("Erro: Entrada inválida. Por favor, digite apenas números.")

# --- Execução do Programa ---
valor_max = 100.0
resultado = solicitar_valor_maximo(valor_max)
print(f"Programa finalizado. Valor salvo: {resultado}")
```

```python
def solicitar_valor():
    valor_maximo = 100

    while True:
        try:
            # Tenta converter a entrada para inteiro
            entrada = input(f"Digite um número até {valor_maximo}: ")
            numero = int(entrada)

            # Verifica se o número está acima do permitido
            if numero > valor_maximo:
                print(f"Erro: O valor {numero} é maior que o máximo permitido ({valor_maximo}).")
            elif numero < 0:
                print("Erro: Por favor, digite um número positivo.")
            else:
                print(f"Sucesso! Você digitou: {numero}")
                break # Sai do loop se a entrada for válida

        except ValueError:
            # Captura erro se o usuário digitar letras ou símbolos
            print("Erro: Entrada inválida. Por favor, digite apenas números inteiros.")

# Executar a função
solicitar_valor()
```

```python
def menu_madeiras():
    # 1. Definição dos tipos de madeira e seus preços
    madeiras = {
        "1": {"nome": "Pinus", "preco": 50.0},
        "2": {"nome": "Eucalipto", "preco": 80.0},
        "3": {"nome": "Cedro", "preco": 120.0},
        "4": {"nome": "Ipê", "preco": 200.0},
        "5": {"nome": "Mogno", "preco": 180.0}
    }

    print("--- MENU DE VENDAS DE MADEIRA ---")
    for chave, valor in madeiras.items():
        print(f"{chave} - {valor['nome']}: R$ {valor['preco']:.2f} (metro linear)")
    print("---------------------------------")

    # 2. Uso de try-except para validar a escolha da madeira
    while True:
        try:
            escolha = input("Digite o código da madeira (1-5): ")

            if escolha not in madeiras:
                print("Erro: Opção inválida. Digite um número de 1 a 5.")
                continue # Volta ao início do while

            # 3. Try-except para validar a quantidade (letras ou números negativos)
            quantidade = float(input(f"Quantos metros de {madeiras[escolha]['nome']} deseja? "))

            if quantidade <= 0:
                print("Erro: A quantidade deve ser maior que zero.")
                continue

            # Cálculo do valor
            total = madeiras[escolha]['preco'] * quantidade
            print(f"\nResumo: {quantidade}m de {madeiras[escolha]['nome']}")
            print(f"Total: R$ {total:.2f}")
            break # Finaliza o loop se tudo estiver correto

        except ValueError:
            # Captura quando o usuário digita letras em vez de números
            print("Erro: Por favor, digite um valor numérico válido.")

# Executar o programa
menu_madeiras()
```

```python
def sistema_vendas():
    # 1. Dicionário com os tipos de madeira e seus preços por unidade
    madeiras = {
        "1": {"nome": "Pinus", "preco": 50.0},
        "2": {"nome": "Eucalipto", "preco": 80.0},
        "3": {"nome": "Cedro", "preco": 150.0},
        "4": {"nome": "Ipê", "preco": 250.0},
        "5": {"nome": "Mogno", "preco": 300.0}
    }

    print("--- Sistema de Vendas de Madeira ---")

    while True:
        # Exibir Menu
        print("\nEscolha o tipo de madeira:")
        for chave, dados in madeiras.items():
            print(f"{chave} - {dados['nome']} (R$ {dados['preco']:.2f})")
        print("0 - Sair")

        try:
            # 2. Captura da escolha (com try/except para letras)
            escolha = input("Digite o número da opção: ")

            if escolha == "0":
                print("Encerrando o sistema. Até logo!")
                break

            if escolha not in madeiras:
                print("ERRO: Opção inválida. Selecione um número de 1 a 5.")
                continue

            # 3. Captura da quantidade (com try/except para letras)
            quantidade = int(input(f"Quantas unidades de {madeiras[escolha]['nome']}? "))

            if quantidade <= 0:
                print("ERRO: A quantidade deve ser um número positivo.")
                continue

            # Cálculo e resultado
            preco_unitario = madeiras[escolha]['preco']
            total = quantidade * preco_unitario

            print(f"\nResumo: {quantidade}x {madeiras[escolha]['nome']}")
            print(f"Total: R$ {total:.2f}")
            print("-" * 30)

        except ValueError:
            # Captura erro se o usuário digitar letras no lugar de números
            print("ERRO: Por favor, digite apenas números inteiros para opção e quantidade.")
        except Exception as e:
            print(f"Ocorreu um erro inesperado: {e}")

# Executar o programa
if __name__ == "__main__":
    sistema_vendas()
```

```python
print("Ola Melque Ferreira Barboza.")
print(" " * 100)
print("Seja bem vindo a Madeireira LASCA VIVA.")
print("__________________SEGUE NOSSO ESTOQUE DE MADEIRA_________________________")

def ESCOLHA_TIPO(): # Aqui vou escolher a madeira.

  while True:
      print("\n Madeira de lei ")
      print("PIN - Pinus (R$ 153.40)")
      print("PER - Peroba (R$ 160.20)")
      print("MOG - Mogno (R$ 190.90)")
      print("IPE - Ipê (R$ 210.10)")
      print("IMB - Imbuia (R$ 220.60)")

      tipo = input("Qual madeira gostaria (PIN/PER/MOG/IPE/IMB): ").upper().strip()

      if tipo == 'PIN': return 153.40
      elif tipo == 'PER': return 160.20
      elif tipo == 'MOG': return 190.90
      elif tipo == 'IPE': return 210.10
      elif tipo == 'IMB': return 220.60
      else:
          # Repetição caso erre algo.
        print("(!) Essa madeira não trabalhamos. Somente do catálago.")

def QTD_TORAS(): # calcula o desconto
    while True:
        try:
          qtd = int(input("\nDigite a quantidade de m³: "))

          if qtd > 2000: # Pedido máximo
            print("(!) Não trabalhamos acima de 2000 m³. Tente novamente.")
            continue
          elif qtd <= 0:
            print("(!) A quantidade deve ser positiva.")
            continue

          if qtd < 100: # Descontos
              desconto = 0.0
          elif 100 <= qtd < 500:
              desconto = 0.04
          elif 500 <= qtd < 1000:
              desconto = 0.09
          elif 1000 <= qtd <= 2000:
              desconto = 0.16
          return qtd, desconto
        except ValueError:
            print("(!) Quantidade inválida. Digite um número inteiro.") # Repete a pergunta caso digite valor não numérico

# Sobre o transporte
def TRANSPORTE():
    while True:
        print("\n--- Escolha o Transporte ---")
        print("1 - Trans. Rodoviário (R$ 1000.00)")
        print("2 - Trans. Ferroviário (R$ 2000.00)")
        print("3 - Trans. Marítimo (R$ 250.00)")

        try:
            opcao = int(input("Escolha a opção (1/2/3): "))
            if opcao == 1:
               return 1000.00
            elif opcao == 2:
               return 2000.00
            elif opcao == 3:
               return 3000.00
            else:
                print("(!) Opção inválida.")
        except ValueError:
            print("(!) Digite um número inteiro para a opção.")

# Funções
valor_unitario = ESCOLHA_TIPO()
quantidade, porcentagem_desconto = QTD_TORAS()
valor_transporte = TRANSPORTE()
subtotal = valor_unitario * quantidade
valor_desconto = subtotal * porcentagem_desconto
total_final = (subtotal - valor_desconto) + valor_transporte

print(" " * 100)
print(f"Valor da madeira: R$ {valor_unitario:.2f}")
print(f"Quantidade: {quantidade}")
print(f"Subtotal: R$ {subtotal:.2f}")
print(f"Desconto ({porcentagem_desconto*100:.0f}%): R$ {valor_desconto:.2f}")
print(f"Transporte: R$ {valor_transporte:.2f}")
print(f"Total Final: R$ {total_final:.2f}")
print(" " * 100)
print("******Obrigado pela preferência, o meio ambiente agradece*******")
```

```python
print("Ola Melque Ferreira Barboza.")
print(" " * 100)

# Variáveis globais
lista_contatos = []
id_global = 5346980

#Funções
def cadastrar_contato(id):
    """Cadastra um novo contato com id único."""
    print("---------------------------------------")
    print(f"MENU CADASTRAR CONTATO - ID: {id}")
    nome = input("Qual nome do contato: ")
    atividade = input("Qual a atividade do contato: ")
    telefone = input("Qual o telefone do contato: ")

    # Armazena em um dicionário
    contato = {
        'id': id,
        'nome': nome,
        'atividade': atividade,
        'telefone': telefone
    }
    lista_contatos.append(contato.copy()) # Copia para lista
    print("Contato cadastrado com sucesso!")

def consultar_contatos():
    while True:
        print(" " * 100)
        print("MENU CONSULTAR CONTATO")
        print("1. Consultar Todos")
        print("2. Consultar por Id")
        print("3. Consultar por Atividade")
        print("4. Retornar ao menu")
        op = input("Escolha a opção desejada: ")

        if op == '1':
            print("Todos os Contatos")
            for contato in lista_contatos:
                print(contato)

        elif opcao == '2':
            id_busca = int(input("Informe o ID do contato: "))
            encontrado = False
            for contato in lista_contatos:
                if contato['id'] == id_busca:
                    print(contato)
                    encontrado = True
                    break
            if not encontrado:
                print("ID não encontrado.")

        elif opcao == '3':
            ativ_busca = input("Informe a atividade: ")
            encontrado = False
            for contato in lista_contatos:
                if contato['atividade'] == ativ_busca:
                    print(contato)
                    encontrado = True
            if not encontrado:
                print("Nenhum contato encontrado com esta atividade.")

        elif opcao == '4':
            return # Retorna ao menu principal
        else:
            print("Opção inválida. Tente novamente.")

def remover_contato():
    print(" " * 100)
    print("MENU REMOVER CONTATO")
    id_remover = int(input("Informe o ID do contato a ser removido: "))

    encontrado = False
    for i, contato in enumerate(lista_contatos):
        if contato['id'] == id_remover:
            lista_contatos.pop(i)
            print("Contato removido com sucesso.")
            encontrado = True
            break

    if not encontrado:
        print("Id inválido. Tente novamente.")

while True:#  (Main)
   print(" " * 100)
   print("MENU PRINCIPAL")
   print("1. Cadastrar Contato")
   print("2. Consultar Contato")
   print("3. Remover Contato")
   print("4. Encerrar Programa")

   op_principal = input("Escolha a opção desejada: ")

   if op_principal == '1':
        id_global += 1
        cadastrar_contato(id_global)
   elif op_principal == '2':
        consultar_contatos()
   elif op_principal == '3':
        remover_contato()
   elif op_principal == '4':
        print("Encerrando programa...")
        break
   else:
        print("Opção inválida. Tente novamente.")
```
